```python
# Hernando Bun - 535230133

import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

iris = pd.read_csv('Iris.csv')
iris.head()
#Membaca file Iris.csv
```

```python
df = iris.drop(['Id', 'Species'], axis =1)
df.head()
#Menampilkan data head dari file Iris
```

```python
plt.plot(df['SepalLengthCm'])
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.show
#Menampilkan data SepalLength
```

```python
plt.plot(df['SepalLengthCm'],color = 'red', marker = '*')
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.show
#Mewarnai Grafik diatas menjadi warna merah
```

```python
plt.plot(df['SepalLengthCm'],'--c')
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.show
# Membuat Grafik diatas menjadi Garis Garis - - -
```

```python
plt.plot(df['SepalLengthCm'],'--c')
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
limits = [ 0, 160, 0, 10]
plt.axis(limits)
plt.show
#menampilkan data Sepal Length dengan format diatas
```

```python
plt.plot(df['SepalLengthCm'])
plt.plot(df['SepalWidthCm'])
plt.plot(df['PetalLengthCm'])
plt.plot(df['PetalWidthCm'])
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.legend(['Sepal Length','Sepal Width','Petal Length', 'Petal Width'])
limits = [ 0, 160, 0, 10]
plt.axis(limits)
plt.show()
#Menampilan 4 grafik dari Iris Data
```

```python
plt.plot(df['SepalLengthCm'])
plt.plot(df['SepalWidthCm'])
plt.plot(df['PetalLengthCm'])
plt.plot(df['PetalWidthCm'])
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.legend(['Sepal Length','Sepal Width','Petal Length', 'Petal Width'], loc ='upper right',ncol =2)
limits = [ 0, 160, 0, 10]
plt.axis(limits)
plt.show()
#Menampilkan data grafik(Sepal Length, Petal Length, Sepal Width, Petal Width) Iris Data
```

```python
c= df.corr()
c
#seberapa pengaruh variabel satu sama lainnya
```

```python
plt.imshow(c, cmap='hot')
plt.show
#sama seperti diatas namun di tampilkan menggunakan heat map
# makin terang korelasi makin tinggi
#makin gelap korelasi makin rendah
```

```python
plt.imshow(c, cmap='hsv')
plt.show()
#sama dengan heatmap diatas semakin merah korelais diatas semakin tinggi
```

```python
ax = sns.heatmap(c, linewidths=0.5)
plt.show
#heatmap menggunakan seaborn 
#lebih jelas karena diperlihatkan SEPAL
```
