```python
moviesClean[
((moviesClean['Year'] >= 2005) &(moviesClean['Year'] <= 2010))&
(moviesClean['Score'] > 8.0) &
(moviesClean['Gross Income'] < moviesClean['Budget'])
]
# Menampilkan  semuafilm yang dirilis antara tahun2005 dan 2010, memiliki rating di atas 8.0, tetapi memiliki pendapatan kurang dari anggaran
```

```python
moviesClean[
(moviesClean['Year'] >= 2010) &
(moviesClean['Score'] > 8.0)  &
(moviesClean['Gross Income'] < moviesClean['Gross Income'].quantile(0.25))
]
#semuafilm yang dirilis setelah 2010  (inklusif), memiliki peringkat di atas 8,0 (eksklusif), tetapi 
#dibuat di bawah persentil ke-25 dalam pendapatan
```
